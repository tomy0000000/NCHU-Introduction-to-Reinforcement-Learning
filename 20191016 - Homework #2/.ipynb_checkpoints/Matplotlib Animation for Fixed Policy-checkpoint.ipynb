## Step 1: Import

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as ani
from matplotlib.patches import Rectangle
from IPython.display import HTML
```

## Step 2: Configure

```python
width = 10
height = 10
blocks = [
    (6, 3),
    (4, 7),
    (8, 2) ]
history = [
    (9, 0),
    (9, 1),
    (9, 2),
    (9, 3),
    (8, 3),
    (7, 3),
    (7, 2),
    (6, 2),
    (5, 2),
    (5, 3),
    (5, 4),
    (5, 5),
    (5, 6),
    (5, 7),
    (5, 8),
    (4, 8),
    (3, 8),
    (3, 7),
    (2, 7),
    (1, 7),
    (0, 7),
    (0, 8),
    (0, 9) ]
```

## Step 3: Initialize Plot

```python
# Grid Shape
fig, ax  = plt.subplots(figsize=(width, height))
xs = np.linspace(0, width, width+1)
ys = np.linspace(0, height, height+1)
ax = plt.gca()

# Grid Boxes
w, h = xs[1] - xs[0], ys[1] - ys[0]
for i, x in enumerate(xs[:-1]):
    for j, y in enumerate(ys[:-1]):
        if (height-j-1, i) in blocks:
            ax.add_patch(Rectangle((x, y), w, h, fill=True, color="#EF476F"))
        elif i % 2 == j % 2:
            ax.add_patch(Rectangle((x, y), w, h, fill=True, color="#06D6A0"))
        else:
            ax.add_patch(Rectangle((x, y), w, h, fill=True, color="#1B9AAA"))

# Grid Lines
for x in xs:
    plt.plot([x, x], [ys[0], ys[-1]], color="#F8FFE5", alpha=.33, linestyle="--")
for y in ys:
    plt.plot([xs[0], xs[-1]], [y, y], color="#F8FFE5", alpha=.33, linestyle="--")

# Grid Point
circle = plt.Circle((0.5, 0.5), 0.25, color="#FFC43D")
ax.add_artist(circle)
```

## Step 4: Initialize Animation

```python
def init_func():
    circle.center = (0.5, 0.5)
    ax.add_patch(circle)
    return circle
def ani_func(i):
    x = history[i][1] + 0.5
    y = height - history[i][0] - 0.5
    circle.center = (x, y)
    ax.add_patch(circle)
    return circle
animate = ani.FuncAnimation(fig, func=ani_func, init_func=init_func, frames=len(history), interval=500)
```

## Step 5: Show Result

```python
HTML(animate.to_jshtml())
```
